```python
cluster_df = pd.read_csv(r'C:\Users\zr_1\final_project\Clustering\clustering_v2.csv')
cluster_df
## Analazying cluster 2
```

```python
cluster_2 = cluster_df[cluster_df['K=8 Cluster Labels'] == 2]
cluster_2
```
