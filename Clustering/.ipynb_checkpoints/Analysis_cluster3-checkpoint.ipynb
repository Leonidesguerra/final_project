```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cluster_df = pd.read_csv('clustering_v2.csv')
```

```python
cluster_3 = cluster_df[cluster_df['K=8 Cluster Labels'] == 3]
cluster_3
```

```python
plt.figure(figsize=(10, 7))
sns.set(style = 'whitegrid')  
sns.stripplot(x="rendimiento", y="estado", jitter=0.05, data=cluster_3)
```

```python
cluster_3.groupby(['K=8 Cluster Labels'])['cultivo_Avena forrajera en verde','cultivo_Frijol',
                              'cultivo_Maíz grano','cultivo_Pastos y praderas',
                              'cultivo_Tomate rojo (jitomate)'].mean()
```

```python
cluster3_cultivo=pd.DataFrame(cluster_3.groupby(['estado'])['cultivo_Pastos y praderas'].sum())
```

```python
cluster3_cultivo
```

```python

cluster_3_g=cluster_3.groupby(['estado'])['cicloproductivo_Otoño-Invierno','cicloproductivo_Perennes',
                              'cicloproductivo_Primavera-Verano','modalidad_Riego',
                              'modalidad_Temporal','rendimiento'].mean()
cluster_3_g
```

```python

cluster_3_g = cluster_3_g.loc[:, (cluster_3_g != 0).any(axis=0)]
```

```python

cluster_3_g
```

```python

cluster3_std=pd.DataFrame(cluster_3.groupby(['estado'])['rendimiento'].std())
cluster3_std
```

```python
cluster_3_final = pd.merge(cluster_3_g, cluster3_std, left_index=True, right_index=True)
```

```python
cluster_3_final.rename(columns= {'rendimiento_x':'rendimiento_promedio','rendimiento_y':'rendimiento_std'},inplace=True)
```

```python
cluster_3_f = pd.merge(cluster_3_final, cluster3_cultivo, left_index=True, right_index=True)
```

```python
cluster_3_f.rename(columns= {'cultivo_Pastos y praderas':'Numero_cultivos'},inplace=True)
```

```python
sns.pairplot(cluster_3_f, height=2.5)
```

```python
plt.figure(figsize=(10, 7))
sns.set(style = 'whitegrid')  
sns.stripplot(x="rendimiento", y=cluster_3_g.index, jitter=0.1, linewidth=1, data=cluster_3_g)
```
