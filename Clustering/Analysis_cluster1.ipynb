```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cluster_df = pd.read_csv('clustering_v2.csv')
```

```python
cluster_1 = cluster_df[cluster_df['K=8 Cluster Labels'] == 1]
cluster_1
```

```python
plt.figure(figsize=(10, 7))
sns.set(style = 'whitegrid')  
sns.stripplot(x="rendimiento", y="estado", jitter=0.05, data=cluster_1)
```

```python
cluster_1.groupby(['K=8 Cluster Labels'])['cultivo_Avena forrajera en verde','cultivo_Frijol',
                              'cultivo_Maíz grano','cultivo_Pastos y praderas',
                              'cultivo_Tomate rojo (jitomate)'].mean()
```

```python
cluster1_cultivo=pd.DataFrame(cluster_1.groupby(['estado'])['cultivo_Maíz grano'].sum())
```

```python
cluster1_cultivo
```

```python

cluster_1_g=cluster_1.groupby(['estado'])['cicloproductivo_Otoño-Invierno','cicloproductivo_Perennes',
                              'cicloproductivo_Primavera-Verano','modalidad_Riego',
                              'modalidad_Temporal','rendimiento'].mean()
cluster_1_g
```

```python

cluster_1_g = cluster_1_g.loc[:, (cluster_1_g != 0).any(axis=0)]
```

```python

cluster_1_g
```

```python

cluster1_std=pd.DataFrame(cluster_1.groupby(['estado'])['rendimiento'].std())
cluster1_std
```

```python
cluster_1_final = pd.merge(cluster_1_g, cluster1_std, left_index=True, right_index=True)
```

```python
cluster_1_final.rename(columns= {'rendimiento_x':'rendimiento_promedio','rendimiento_y':'rendimiento_std'},inplace=True)
```

```python
cluster_1_f = pd.merge(cluster_1_final, cluster1_cultivo, left_index=True, right_index=True)
```

```python
cluster_1_f.rename(columns= {'cultivo_Maíz grano':'Numero_cultivos'},inplace=True)
```

```python
sns.pairplot(cluster_1_f, height=2.5)
```

```python
plt.figure(figsize=(10, 7))
sns.set(style = 'whitegrid')  
sns.stripplot(x="rendimiento", y=cluster_1_g.index, jitter=0.1, linewidth=1, data=cluster_1_g)
```
