```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cluster_df = pd.read_csv('clustering_v2.csv')
```

```python
cluster_4 = cluster_df[cluster_df['K=8 Cluster Labels'] == 4]
cluster_4
```

```python
cluster_4["rendimiento"].mean()
```

```python
plt.figure(figsize=(10, 7))
sns.set(style = 'whitegrid')  
sns.stripplot(x="rendimiento", y="estado", jitter=0.05, data=cluster_4)
```

```python
cluster_4.groupby(['K=8 Cluster Labels'])['cultivo_Avena forrajera en verde','cultivo_Frijol',
                              'cultivo_Maíz grano','cultivo_Pastos y praderas',
                              'cultivo_Tomate rojo (jitomate)'].mean()
```

```python
cluster4_cultivo=pd.DataFrame(cluster_4.groupby(['estado'])['cultivo_Tomate rojo (jitomate)'].sum())
```

```python
cluster4_cultivo
```

```python

cluster_4_g=cluster_4.groupby(['estado'])['cicloproductivo_Otoño-Invierno','cicloproductivo_Perennes',
                              'cicloproductivo_Primavera-Verano','modalidad_Riego',
                              'modalidad_Temporal','rendimiento'].mean()
cluster_4_g
```

```python

cluster_4_g = cluster_4_g.loc[:, (cluster_4_g != 0).any(axis=0)]
```

```python

cluster_4_g
```

```python

cluster4_std=pd.DataFrame(cluster_4.groupby(['estado'])['rendimiento'].std())
cluster4_std
```

```python
cluster_4_final = pd.merge(cluster_4_g, cluster4_std, left_index=True, right_index=True)
```

```python
cluster_4_final.rename(columns= {'rendimiento_x':'rendimiento_promedio','rendimiento_y':'rendimiento_std'},inplace=True)
```

```python
cluster_4_f = pd.merge(cluster_4_final, cluster4_cultivo, left_index=True, right_index=True)
```

```python
cluster_4_f.rename(columns= {'cultivo_Tomate rojo (jitomate)':'Numero_cultivos'},inplace=True)
```

```python
sns.pairplot(cluster_4_f, height=2.5)
```

```python
plt.figure(figsize=(10, 7))
sns.set(style = 'whitegrid')  
sns.stripplot(x="rendimiento", y=cluster_4_g.index, jitter=0.1, linewidth=1, data=cluster_4_g)
```
