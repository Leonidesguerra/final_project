```python
from sqlalchemy import create_engine
import psycopg2
```

```python
# create engine
engine = create_engine(DATABASE_URI)

#Postgres connection
#rds_connection_string = "postgres:{pwd}@localhost:5432/Agriculture_JB"
#engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# convert the dataframe into a sql table with pd.to_sql function
df.to_sql(con= engine, name='clustering', index=False, if_exists = "replace")
```

```python
data = engine.execute(
        "SELECT latitud, longitud, cultivo, estado, clusters ,rendimiento  FROM clustering;")
all_data = []
for record in data:
    data_dict = {}
    data_dict['latitud'] = record[0]
    data_dict['longitud'] = record[1]
    data_dict['cultivo'] = record[2]
    data_dict['estado'] = record[3]
    data_dict['clusters'] = record[4]
    data_dict['rendimiento'] = record[5]
    all_data.append(data_dict)
all_data
```
