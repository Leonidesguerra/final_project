```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Analizing Cluster 5

```python
cluster_df = pd.read_csv('clustering_v2.csv')
```

```python
cluster_6 = cluster_df[cluster_df['K=8 Cluster Labels'] == 6]
cluster_6
```

```python
cluster_6['rendimiento'].mean()
```

```python
plt.figure(figsize=(10, 7))
sns.set(style = 'whitegrid')  
sns.stripplot(x="rendimiento", y="estado", jitter=0.05, data=cluster_6)
```

### Group the dataframe to identify which is the predominant crop 

```python
cluster_6.groupby(['K=8 Cluster Labels'])['cultivo_Avena forrajera en verde','cultivo_Frijol',
                              'cultivo_Maíz grano','cultivo_Pastos y praderas',
                              'cultivo_Tomate rojo (jitomate)'].mean()
```

#### We can see that for the cluster 5 the principal crop is 'cultivo_Pastos y praderas	'

```python
cluster6_cultivo=pd.DataFrame(cluster_6.groupby(['estado'])['cultivo_Pastos y praderas'].sum())
```

```python
cluster6_cultivo
```

```python
cluster_6_g=cluster_6.groupby(['estado'])['cicloproductivo_Otoño-Invierno','cicloproductivo_Perennes',
                              'cicloproductivo_Primavera-Verano','modalidad_Riego',
                              'modalidad_Temporal','rendimiento'].mean()
cluster_6_g
```

```python
cluster_6_g = cluster_6_g.loc[:, (cluster_6_g != 0).any(axis=0)]
```

```python
cluster_6_g
```

```python
cluster6_std=pd.DataFrame(cluster_6.groupby(['estado'])['rendimiento'].std())
cluster6_std
```

```python
cluster_6_final = pd.merge(cluster_6_g, cluster6_std, left_index=True, right_index=True)
```

```python
cluster_6_final.rename(columns= {'rendimiento_x':'rendimiento_promedio','rendimiento_y':'rendimiento_std'},inplace=True)
```

```python
cluster_6_f = pd.merge(cluster_6_final, cluster6_cultivo, left_index=True, right_index=True)
```

```python
cluster_6_f.rename(columns= {'cultivo_Pastos y praderas':'Numero_cultivos'},inplace=True)
```

```python
sns.pairplot(cluster_6_f, height=2.5)
```

```python
plt.figure(figsize=(10, 7))
sns.set(style = 'whitegrid')  
sns.stripplot(x="rendimiento", y=cluster_6_g.index, jitter=0.1, linewidth=1, data=cluster_6_g)
```
